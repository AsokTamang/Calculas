```python
def gradient_descent(x,y,dfdx,dfdy,learning_rate=0.1,iterations=100):
    for i in range(iterations):
        x,y = x - learning_rate * dfdx(x,y) , y - learning_rate * dfdy(x,y)
    return x,y    
```
