```python
import numpy as np
```

```python
def f_example_1(x): #this is our function f(x)
    return np.exp(x) - np.log(x)

def dfdx_example_1(x):  #this is the derivative of that function
    return np.exp(x) - 1/x
```

```python
def gradient_descent(dfdx,x,learning_rate=0.1,iterations=100): #here x is the initial value or point in a curve,dfdx is the derivative function
    for i in range(iterations):
        x = x - (learning_rate * dfdx(x))  #x keeps on changing until the iterations stop
    return x  #this is the new value , which might be the minimal of the function 
    
    
```

```python
x_initial=1.6;learning_rate=0.1;iterations=25
print('The gradient descent of the function f_example_1 is:',gradient_descent(dfdx_example_1,x_initial,learning_rate,iterations))
```
